```python
import json
from pathlib import Path

log_dir = Path(r"d:\amir lab\RS-Vamba-main\fusion\outputs\logs")

with open(log_dir / "history.json") as f:
    history = json.load(f)

# find the last time epoch counter reset to 1 -> that's where clean retrain starts
start_idx = max(i for i, h in enumerate(history) if h['epoch'] == 1)
new_history = history[start_idx:]

print(f"Keeping {len(new_history)} epochs from clean retrain:")
for h in new_history:
    print(f"  Epoch {h['epoch']} | val_acc: {h['val_acc']:.4f}")

with open(log_dir / "history.json", "w") as f:
    json.dump(new_history, f, indent=2)

print("\nHistory cleaned and saved.")
```
